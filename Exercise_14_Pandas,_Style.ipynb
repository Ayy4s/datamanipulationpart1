```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12],
    'D': [13, 14, 15, 16]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 6))
```

```python
colors = ["#FFFFFF", "#E9F6F7", "#A3E4F7", "#70D8F0", "#3AC1E9", "#0396C2", "#01708A", "#004752"]
sns.set_palette(sns.color_palette(colors))
```

```python
sns.heatmap(df, annot=True, cmap="YlGnBu", linewidths=0.5, fmt='g')
```

```python
plt.title('DataFrame Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
