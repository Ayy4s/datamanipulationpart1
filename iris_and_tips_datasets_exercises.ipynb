```python
#1-Load the 'tips' dataset from seaborn and display the first 5 rows.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
print(tips.head(5))
```

```python
# 2-Add a column 'tip_percentage' that shows the tip as a percentage of the total bill.
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100
tips.head()
```

```python
#3-Visualize the relationship between 'total_bill' and 'tip' using a scatter plot. Use 'time' as hue.
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title("Relationship between Total Bill and Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
```

```python
#4-Calculate the average 'tip_percentage' for each 'day'.
av_tip_per= tips.groupby('day')['tip_percentage'].mean()
av_tip_per
```

```python
#5-Visualize 'average_tip_percentage' for each 'day' using a bar plot.
sns.barplot(x=av_tip_per.index,y=av_tip_per.values)
```

```python
#6-Create a box plot of 'total_bill' for each 'day'.
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
#7-Create a new column 'bill_per_person' which is the ratio of total bill to size

tips['bill_per_person'] = tips['total_bill'] / tips['size']
tips.head()
```

```python
#8-Visualize 'bill_per_person' for each 'day' using a box plot.
sns.boxplot(data=tips, x='day', y='bill_per_person')
```

```python
#9-Create a correlation matrix for the 'tips' dataset and visualize it using a heatmap.
corr=tips.corr()
sns.heatmap(corr , annot=True)
```

```python
#10-Plot a pairplot of the numerical features in the 'tips' dataset, colored by 'sex'.
sns.pairplot(tips[(['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size','tip_percentage','bill_per_person'])],hue='sex')
```

```python
#11-Load the 'iris' dataset from seaborn and display the first 5 rows.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
iris=sns.load_dataset('iris')
iris.head()
```

```python
#12-Create a function to categorize each flower by petal length: 'small' if petal length is less than or equal to 2, 'medium' if petal length is between 2 and 5, 'large' if petal length is greater than 5. Apply this function to create a new column 'petal_size'.
iris['petal_size']=pd.cut(iris['petal_length'],bins=[0,2,5,iris['petal_length'].max()],labels=["Small","Medium","Large"])
iris.head()
```

```python
#13-Using the lambda function, create a new column 'sepal_area' which is the product of 'sepal_length' and 'sepal_width'.
iris['sepal_area'] = iris.apply(lambda row: row['sepal_length'] * row['sepal_width'], axis=1)
print(iris.head())
```

```python
#14-Compute the mean 'sepal_area' for each 'species'.
mean_sepal_area = iris.groupby('species')['sepal_area'].mean()
print(mean_sepal_area)
```

```python
#15-Convert the 'species' column to categorical type and print the datatype of each column in the dataset.
iris['species'] = pd.Categorical(iris['species'])
print(iris.dtypes)
```

```python
#16-Create a new DataFrame from the 'iris' dataset that only contains the 'sepal_area' and 'petal_size' columns.
new_iris = iris[['sepal_area', 'petal_size']].copy()

new_iris.head()
```

```python
#17-Rename all columns in the dataset so that they don't contain underscores.
new_iris.rename(columns = {'sepal_area':'sepal area', 'petal_size':'petal size'}, inplace=True)

new_iris.head()
```

```python
#18-Apply a function to the 'species' column that replaces 'setosa' with 'new_species', keep the other species' names unchanged.
iris['species'].replace({'setosa' : 'new_species'}, inplace=True)
iris.head()
```

```python
#19-Drop the 'sepalarea' and 'petalsize' columns from the dataset.
iris = iris.drop(['sepal_area', 'petal_size'], axis=1)
iris.head()
```
