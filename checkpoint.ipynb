```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['petal_size']=pd.cut(iris['petal_length'],bins=[0,2,5,iris['petal_length'].max()],labels=["Small","Medium","Large"])
iris.head()
```

```python
iris['sepal_area'] = iris.apply(lambda x: x['sepal_length'] * x['sepal_width'], axis=1)
```

```python
iris.groupby('species')['sepal_area'].mean()
```
